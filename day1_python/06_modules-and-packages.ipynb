# Concepts about modules

- Can contain 
	- executable statements 
	- function definitions. 
- executed 
	- once when imported
	- or when run as a script
- One module can import other modules
- import into current module's **symbol table**

	> from functions import sum
	
	> sum(1, 2)

- import all names (**bad idea**)
	> from fibo import *

- Each module is only imported once per interpreter session. 
	- Restart interpreter if you know module has changed
	- or importlib.reload(modulename)

- When you run a Python module, code will be executed
	- \_\_name\_\_: \_\_main\_\_


# demonstrate \_\_main\_\_ running in script 

- The Module Search Path by interpreter
	- built-in module with that name
	- a file named modulename.py in sys.path, which is initialized as follows
		- The directory containing the input script or the current directory
		- PYTHONPATH (directory names, syntax as the shell variable PATH).
		- The installation-dependent default.

- Programs can modify sys.path

# “Compiled” Python files

- __pycache__ directory
- Checks modification date of the source against the compiled version for recompilation
- Compiled versions are platform independent
- only loading is faster. Runs with same speed

# Packages

- Organize modules into packages
- Keep module files inside directories

sound/                       Top-level package
      __init__.py            Initialize the package
      formats/               Subpackage for file format 
              __init__.py
              wavread.py
              wavwrite.py
              aiffread.py
              aiffwrite.py
              auread.py
              auwrite.py
              ...
      effects/                  Subpackage for sound effects
              __init__.py
              echo.py
              surround.py
              reverse.py
              ...
      filters/                  Subpackage for filters
              __init__.py
              equalizer.py
              vocoder.py
              karaoke.py
              ...

- \_\_init\_\_.py tells python that this is a package
- Mostly empty but can run initialization code

> import sound.effects.echo

> sound.effects.echo.echofilter(input, output, delay=0.7, atten=4)

> from sound.effects import echo

> echo.echofilter(input, output, delay=0.7, atten=4)

> from sound.effects.echo import echofilter

> echofilter(input, output, delay=0.7, atten=4)

- import *
	- \_\_init\_\_.py: \_\_all\_\_
	> __all__ = ["echo", "surround", "reverse"]


- Relative imports: relative to the current module

from . import echo
from .. import formats
